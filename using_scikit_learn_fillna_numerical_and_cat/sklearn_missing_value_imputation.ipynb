```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
import seaborn as sns
```

```python
train_data = pd.read_csv('/content/house_prince_train_data.csv')
test_data = pd.read_csv('/content/housse_prince_test_datta.csv')
```

```python
train_data.head()
```

```python
x_train=train_data.drop(columns=['SalePrice'])
y_train=train_data['SalePrice']
```

```python
x_train.shape
```

```python
y_train.shape
```

# numerical missing value implutation

```python
# select numerical data
num_var=x_train.select_dtypes(include=['int64','float64']).columns
```

```python
num_var
```

```python
missing_value=x_train[num_var].isnull().sum()
```

```python
missing_value
```

```python
numeri_imputer=SimpleImputer(strategy='mean')
numeri_imputer
```

```python
numeri_imputer.fit(x_train[num_var]) # fit data
```

```python
numeri_imputer.statistics_
```

```python
x_train[num_var]=numeri_imputer.transform(x_train[num_var])
test_data[num_var]=numeri_imputer.transform(test_data[num_var])

```

```python
x_train[num_var].isnull().sum()
```

# cat_variblees missing value imputation

```python
cat_var=x_train.select_dtypes(include=['object']).columns
cat_var
```

```python
x_train[cat_var].isnull().sum()
```

```python
impute_cat=SimpleImputer(strategy="most_frequent")
impute_cat.fit(x_train[cat_var])
```

```python
x_train[cat_var]=impute_cat.transform(x_train[cat_var])
test_data[cat_var]=impute_cat.transform(test_data[cat_var])
```
